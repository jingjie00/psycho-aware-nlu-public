```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
from IPython.display import display
import pickle
```

```python
mbti_path = "./mbti.csv"
mbti_corpus = pd.read_csv(mbti_path)
```

# Part 3: Organise corpus

```python
mbti_corpus.columns = ["user", "type","text"]
mbti_corpus['O'] = np.select(list(map(mbti_corpus['type'].str.contains, ['S','N'])), [0,1])
mbti_corpus['C'] = np.select(list(map(mbti_corpus['type'].str.contains, ['P','J'])), [0,1])
mbti_corpus['E'] = np.select(list(map(mbti_corpus['type'].str.contains, ['I','E'])), [0,1])
mbti_corpus['A'] = np.select(list(map(mbti_corpus['type'].str.contains, ['T','F'])), [0,1])
for i,ed in mbti_corpus.iterrows():
    mbti_corpus.at[i,"ptype"] = str(int(str(ed["O"])+str(ed["C"])+str(ed["E"])+str(ed["A"]),2))
mbti_corpus['user'] = mbti_corpus.index
mbti_corpus = mbti_corpus[["user", "O", "C","E", "A","ptype", "text"]]
```

# Part 5: Analyse and Save

# Visualization

```python

mbti_dimension=["O","C","E","A"]

def tabulate_baseline(df):
  df = df.astype('object').describe()
  df = df.T
  df["baseline"] = df["freq"]/df["count"]
  display(df)
  df = df[["baseline"]]
  df = df.T

def analyse(corpus):
    corpus["ptype"].value_counts()
    tabulate_baseline(corpus)
```

```python

print("MBTI corpus")
analyse(mbti_corpus)
print("[End]========================")
```

# Save

```python
def save_set(corpus_name,corpora):
    dir = "../../dataset/"
    if not os.path.exists(dir):
        os.makedirs(dir)
    pickle.dump(corpora, open(dir+"/"+corpus_name+".pickle", 'wb'))
```
