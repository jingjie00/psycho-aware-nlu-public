```python
# load pandas pickle file
import pandas as pd
import numpy as np
import pickle
import os
def extract(path):
    print(os.getcwd())
    file = open(path, 'rb')
    object_file = pickle.load(file)
    file.close()
    return object_file

def save_dataset(item, dir, name):
    if not os.path.exists(dir):
        os.makedirs(dir)
    path = dir+"/"+name+".pickle"
    pickle.dump(item, open(path, 'wb'))
```

```python
import pickle
imdb_p = pickle.load(open("../../corpus/personality-aware-sentiment/albert-base-list-imdb.pickle", "rb"))

imdb_p = imdb_p[['review', 'highest', 'sentiment']]
imdb_p['E'] = imdb_p['highest'].str.contains('E').astype(int)
imdb_p['F'] = imdb_p['highest'].str.contains('F').astype(int)
imdb_p['N'] = imdb_p['highest'].str.contains('N').astype(int)
imdb_p['J'] = imdb_p['highest'].str.contains('J').astype(int)
imdb_p.drop("highest", axis=1, inplace=True)


imdb_e = pickle.load(open("../../corpus/emotion-aware-sentiment/roberta-base-imdb.pickle", "rb"))

imdb_e.drop("review", axis=1, inplace=True)
imdb_e.drop("sentiment", axis=1, inplace=True)

imdb = pd.concat([imdb_p, imdb_e], axis=1)
#remove duplicate

save_dataset(imdb, "../../corpus/mixed", "imdb")

imdb.head()
```

```python
import pickle
moviereview_p = pickle.load(open("../../corpus/personality-aware-sentiment/albert-base-list-moviereview.pickle", "rb"))

moviereview_p = moviereview_p[['content', 'label','highest']]
moviereview_p['E'] = moviereview_p['highest'].str.contains('E').astype(int)
moviereview_p['F'] = moviereview_p['highest'].str.contains('F').astype(int)
moviereview_p['N'] = moviereview_p['highest'].str.contains('N').astype(int)
moviereview_p['J'] = moviereview_p['highest'].str.contains('J').astype(int)
moviereview_p.drop("highest", axis=1, inplace=True)

moviereview_e = pickle.load(open("../../corpus/emotion-aware-sentiment/roberta-base-moviereview.pickle", "rb"))

moviereview_e.drop("content", axis=1, inplace=True)
moviereview_e.drop("label", axis=1, inplace=True)

moviereview = pd.concat([moviereview_p, moviereview_e], axis=1)
#remove duplicate

save_dataset(moviereview, "../../corpus/mixed", "moviereview")

moviereview.head()
```
